```python
# prompt: can you find out mean, median, mode, standard deviation, variance for the given data

import pandas as pd

# Load the dataset (assuming 'Salaries.csv' is in the current directory)
df = pd.read_csv('Salaries.csv')

# Calculate mean, median, mode, standard deviation, and variance for the 'salary' column
mean_salary = df['salary'].mean()
median_salary = df['salary'].median()
mode_salary = df['salary'].mode()
std_salary = df['salary'].std()
var_salary = df['salary'].var()

# Print the results
print(f"Mean salary: {mean_salary}")
print(f"Median salary: {median_salary}")
print(f"Mode salary: {mode_salary}")
print(f"Standard deviation of salary: {std_salary}")
print(f"Variance of salary: {var_salary}")
```

```python
# prompt: plot a bar graph of SAT column from the above data

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df1 = pd.read_csv('Universities.csv')

# Create the bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size for better visualization
plt.bar(df1.index, df1['SAT'])
plt.xlabel('University Index')
plt.ylabel('SAT Score')
plt.title('SAT Scores of Universities')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for readability
plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
